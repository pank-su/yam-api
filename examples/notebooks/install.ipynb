# Установка библиотеки

Библиотека, пока не выложена, так, что для установки необходимо собрать её локально:

```shell
git clone https://github.com/pank-su/yam-api
cd yam-api
./gradlew publishToMavenLocal
```

После этого вы можете установить библиотеку в зависимости через gradle:

```kotlin
repositories{
    // ... Ваши репозитории
    mavenLocal()
}

dependencies{
    // Ваши зависимости
    implementation("su.pank.yamapi:core:1.0.0-dev")
    // Не забудьте добавить ktor engine для вашей платформы https://ktor.io/docs/client-engines.html
}
```

Если библиотека установится правильно, то вы сможете запустить следующий код:

````kotlin
import kotlinx.coroutines.runBlocking
import su.pank.yamapi.builder.createYaMusicApiClient

val client = createYaMusicApiClient {}

runBlocking {
    client.account.status()
}
````
