# Получение токена

Для того чтобы получить токен воспользуйтесь методами описанными по ссылке: https://yandex-music.readthedocs.io/en/main/token.html

Если ваш токен корректный, то этот код должен сработать без ошибки:

```kotlin
import kotlinx.coroutines.runBlocking
import su.pank.yamapi.builder.createYaMusicApiClient
import su.pank.yamapi.exceptions.NotAuthenticatedException

val myToken = "" // Вставьте сюда ваш токен

val client = createYaMusicApiClient {
    token = myToken
}

// Получаем ID аккаунта, если он null, то выбрасываем ошибку
runBlocking {
    client.account.status().account.uid ?: throw NotAuthenticatedException()
}
```
